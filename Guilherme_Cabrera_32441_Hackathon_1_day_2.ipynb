```python
payment = 10000.00
debts = {"BMW":2500.45,"Apartamento":3500.5,"Veterinario":500,"Luz":100.80,"Água":80.35,"Condominio":800.40,"Gás":150.65}
def getNetPayment(payment,debts):
  print("*"*100)
  print(f"Seu salário é de {payment} e custos a seguir")
  totalCost = 0
  for cost in debts:
    print(f"Parcela da {cost} = {debts[cost]}")
    totalCost += debts[cost]
  print("*"*100)
  print("O valor disponível é de {:.2f}".format(payment-totalCost))
    
getNetPayment(payment,debts)

print("\n")
print("*"*100)
print(f"Seu salário é de cientista de dados jr é de{payment} e seus gastos estão dispostos a seguir\n"+
      f"Parcela da BMW = {debts['BMW']}\n"+
      f"Parcela do aplarmento = {debts['Apartamento']}\n"+
      f"Veterinario do Romeuzinho = {debts['Veterinario']}\n"+
      f"Conta de luz = {debts['Luz']}\n" +
      f"Conta de água: {debts['Água']}\n"+
      f"Conta de condomínio = {debts['Condominio']}\n"+
      f"Conta de gás = {debts['Gás']}")
print("*"*100)
print("O valor disponível para colocar na popança é de: {:.2f}".format((debts['BMW']+debts['Apartamento']+debts['Veterinario']+debts['Luz']+debts['Água']+debts['Condominio']+debts['Gás'])-payment))
```
